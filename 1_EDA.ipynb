```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Plot style
sns.set(style="whitegrid")

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd

# Update the path if needed
df = pd.read_csv("/home/shruti/Desktop/creditcard.csv")

df.head()
```

```python
print("Dataset Shape:", df.shape)
df.info()
```

# The dataset contains 284,807 transactions with 31 features. All features are numerical and there are no missing values, indicating that the dataset is already clean and suitable for machine learning analysis.

```python
df.describe()
```

# Most features are PCA-transformed with mean values close to zero. The Amount feature shows a wide range and is not standardized, indicating the need for feature scaling during preprocessing.

```python
df["Class"].value_counts()
df["Class"].value_counts(normalize=True) * 100
```

# Fraudulent transactions account for only about 0.17% of the dataset, indicating a highly imbalanced classification problem.

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Class", data=df)
plt.title("Distribution of Fraud vs Legitimate Transactions")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

# The severe class imbalance implies that accuracy alone is not an appropriate evaluation metric. Special techniques are required to handle imbalance during model training.

```python
fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

print("Fraud transactions:", fraud.shape[0])
print("Normal transactions:", normal.shape[0])
```

```python
plt.figure(figsize=(10,5))

sns.histplot(normal["Amount"], bins=50, color="blue", label="Normal", stat="density")
sns.histplot(fraud["Amount"], bins=50, color="red", label="Fraud", stat="density")

plt.legend()
plt.xlabel("Transaction Amount")
plt.ylabel("Density")
plt.title("Transaction Amount Distribution: Fraud vs Normal")
plt.show()
```

# Fraudulent transactions occur across a wide range of transaction amounts, including both low- and high-value transactions. This indicates that transaction amount alone is insufficient for detecting fraud and must be combined with other features.

```python
plt.figure(figsize=(12,4))
plt.plot(df["Time"], df["Amount"], alpha=0.3)
plt.xlabel("Time (seconds since first transaction)")
plt.ylabel("Transaction Amount")
plt.title("Transaction Amounts Over Time")
plt.show()
```

# Transactions are distributed continuously over time without obvious periodicity. Fraudulent transactions do not occur at fixed intervals, highlighting the need for feature-based rather than time-based detection.

```python
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", linewidths=0.2)
plt.title("Feature Correlation Heatmap")
plt.show()
```

# Most PCA-transformed features show low pairwise correlation, while some features exhibit stronger relationships with the target variable. This suggests that nonlinear models may be effective for fraud detection.
